```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(0)
data = np.random.rand(10, 1)


# Generate target
target = 2 * data + 1 + 0.1 * np.random.randn(10, 1)

# Add bias
data_with_bias = np.hstack((np.ones((10, 1)), data))  # Adding a column of ones for bias

# Calculate weights without bias
weights_without_bias = np.linalg.inv(data.T @ data) @ data.T @ target

# Calculate weights with bias
weights_with_bias = np.linalg.inv(data_with_bias.T @ data_with_bias) @ data_with_bias.T @ target


plt.scatter(data, target, label='Data Points')
plt.plot(data, data @ weights_without_bias, label='Regression Line (Without Bias)', color='red')
plt.plot(data, data_with_bias @ weights_with_bias, label='Regression Line (With Bias)', color='green')
plt.xlabel('Data')
plt.ylabel('Target')
plt.legend()
plt.title('Linear Regression with and without Bias')
plt.show()

print("Weights without bias:")
print(weights_without_bias)

print("\nWeights with bias:")
print(weights_with_bias)
```

```python
# Test data
test_data_1 = 0.3
test_data_2 = 0.7

# Predict target values using the line without bias
prediction_without_bias_1 = test_data_1 * weights_without_bias
prediction_without_bias_2 = test_data_2 * weights_without_bias

# Predict target values using the line with bias
test_data_with_bias_1 = np.array([1, test_data_1])
test_data_with_bias_2 = np.array([1, test_data_2])

prediction_with_bias_1 = test_data_with_bias_1 @ weights_with_bias
prediction_with_bias_2 = test_data_with_bias_2 @ weights_with_bias

print("Predictions without bias:")
print(f"Test Data 1: {prediction_without_bias_1}")
print(f"Test Data 2: {prediction_without_bias_2}")

print("\nPredictions with bias:")
print(f"Test Data 1: {prediction_with_bias_1}")
print(f"Test Data 2: {prediction_with_bias_2}")
```

```python
new_data = np.array([[0.2], [0.8]])
new_target = 2 * new_data + 1 + 0.1 * np.random.randn(2, 1)

# Add new data
data = np.vstack((data, new_data))
target = np.vstack((target, new_target))

# Calculate weights without bias
weights_without_bias = np.linalg.inv(data.T @ data) @ data.T @ target

# Add bias
data_with_bias = np.hstack((np.ones((12, 1)), data))

# Calculate weights with bias
weights_with_bias = np.linalg.inv(data_with_bias.T @ data_with_bias) @ data_with_bias.T @ target


plt.scatter(data, target, label='Data Points (Extended)')
plt.plot(data, data @ weights_without_bias, label='Regression Line (Without Bias, Extended)', color='blue')
plt.plot(data, data_with_bias @ weights_with_bias, label='Regression Line (With Bias, Extended)', color='purple')
plt.xlabel('Data')
plt.ylabel('Target')
plt.legend()
plt.title('Linear Regression with and without Bias (Extended Data)')
plt.show()

print("Weights without bias (Extended Data):")
print(weights_without_bias)

print("\nWeights with bias (Extended Data):")
print(weights_with_bias)
```

```python
# Test data
test_data_1 = 0.3
test_data_2 = 0.7

# Predict target values using the line without bias for the extended dataset
prediction_without_bias_1 = test_data_1 * weights_without_bias
prediction_without_bias_2 = test_data_2 * weights_without_bias

# Predict target values using the line with bias for the extended dataset
test_data_with_bias_1 = np.array([1, test_data_1])
test_data_with_bias_2 = np.array([1, test_data_2])

prediction_with_bias_1 = test_data_with_bias_1 @ weights_with_bias
prediction_with_bias_2 = test_data_with_bias_2 @ weights_with_bias

print("Predictions without bias (Extended Data):")
print(f"Test Data 1: {prediction_without_bias_1}")
print(f"Test Data 2: {prediction_without_bias_2}")

print("\nPredictions with bias (Extended Data):")
print(f"Test Data 1: {prediction_with_bias_1}")
print(f"Test Data 2: {prediction_with_bias_2}")
```
